# **Loading Embedding Model**

```python
from sentence_transformers import SentenceTransformer
embed_model = SentenceTransformer("thenlper/gte-large")
```

```python
def get_embedding(text: str) -> list[float]:
  if not text.strip():
    print("Attempted to get embedding for empty text.")
    return []
  embeddings = embed_model.encode(text)
  return embeddings.tolist()
```

```python
data["embeddings"] = data["fullplot"].apply(get_embedding)
```

# **Retrieval**

```python
# Set this info in MongoDB Atlas Search
# {
#     "fields": [{
#         "numDimensions": 1024,
#         "path": "embedding",
#         "similarity": "cosine",
#         "type": "vector"
#     }]
# }
```

```python
def vector_search(user_query, collection):
  query_embedding = get_embedding(user_query)
  print(query_embedding)

  if query_embedding is None:
    return "Invalid query or embedding is failed"

  pipeline = [
      {
          "$vectorSearch": {
              "index": "vector_index",
              "queryVector": query_embedding,
              "path": "embedding",
              "numCandidates": 150,
              "limit": 4
          }
      },
      {
          "$project": {
              "fullplot": 1,
              "title": 1,
              "genres": 1,
              "score": {"$meta": "vectorSearchScore"}
          }
      },
  ]
  result = collection.aggregate(pipeline)
  return list(result)
```

```python
vector_search("What is the best horror movie to watch and why ?", collection)
```

```python
query = "What is the best horror movie to watch and why ?"
```

```python
def get_search_result(query, collection):
  get_knowledge = vector_search(query, collection)
  search_result = ""

  for result in get_knowledge:
    search_result += "Title: {0}, Plot: {1}".format(result.get("title", "N/A"), result.get("fullplot", "N/A"))

  return search_result
```

```python
source_information = get_search_result(query, collection)
```
